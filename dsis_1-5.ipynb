Изучение matplotlib.pyplot

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 7, 15, 24, 32]
#plt.plot() - функция строит график в соответствии со значениями, которые ей были переданы (координаты точек)
plt.plot(x,y, color = "pink", marker = '*', markersize = 10) #8, >< <, ^, o, d, D, H, h и др могут быть маркерами
#подписи для осей
plt.xlabel("Время")
plt.ylabel("Скорость")
plt.title("График №1")
plt.figure(figsize = (5, 5))
#plt.show() - визуализированный вывод данных
plt.show()
```

```python
import numpy as np
```

```python
x = np.random.random(100)
y = np.random.random(100)
sizes = np.random.randint(7,95,100)
colors = np.random.random(100)
plt.scatter(x, y, s = sizes, c = colors, marker = '*')
plt.title("График №2")
plt.show()
```

```python
x = ['Январь', 'Февраль', 'Март', 'Апрель', 'Май']
y = [2, 4, 3, 1, 7]
#alpha - прозрачность
plt.bar(x, y, label='Величина прибыли', color = 'maroon', alpha = 0.5) #Параметр label позволяет задать название величины для легенды
plt.xlabel('Месяц года')
plt.ylabel('Прибыль, в млн руб.')
plt.title('График №3')
plt.legend() #окошко в левом углу
plt.show()
```

```python
counts = [34, 15, 76, 82]
names = ["Ford", "Toyota", "BMW", "Audi"]

plt.pie(counts, labels = names, autopct = '%1.1f%%') #autopct - значение в процентах, '%1.1f%%' - один знак после запятой
plt.title('График №4')
plt.show()
```

Изучение seaborn

```python
import seaborn as sns
import pandas as pd
```

```python
info = sns.load_dataset("iris") #встроенная бд
plt.figure(figsize = (10, 5))
sns.lineplot(x="sepal_length", y = "sepal_width", data = info)
plt.title('График №5')
plt.show()
```

```python
flights_data = sns.load_dataset("flights")
plt.title('Таблица №1')
flights_data.head()
```

```python
sns.scatterplot(data=flights_data, x="year", y="passengers")
plt.title('График №6')
plt.show()
```

```python
sns.barplot(data=flights_data, x="year", y="passengers")
plt.title('График №7')
plt.show()
```

```python
diamonds_data = sns.load_dataset('diamonds')
plt.subplot(1, 2, 1)
sns.countplot(x='carat', data=diamonds_data)
plt.subplot(1, 2, 2)
sns.countplot(x='depth', data=diamonds_data)
plt.title('График №8')
plt.show()
#subplot принимает в себя три параметра: первый – количество строк, второй – количество столбцов, третий – количество диаграмм.
```
